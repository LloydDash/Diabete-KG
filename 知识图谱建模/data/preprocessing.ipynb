```python
!pip install xlrd
drug_data = pd.read_excel("drug.xls",skiprows=[0])
print(drug_data)
```

提取降糖药名称

```python
drug = drug_data["药 品 名 称"]
print(drug)
```

```python
print(type(drug))
```

```python
# transform pandas.core.series.Series to list of strings
drug_list = []
for i in drug:
    # 去除float型空值
    if str(i)=='nan':
        continue

    # 用是否含括号（别名）区分药物名称和种类描述
    # 英文括号()
    if i.find('服药时间')>=0:
        continue
    if i.find('(')>=0 and i.find(')')>=0:
        print(i,i.find('('),i.find(')'))
        drug_list.append(i)
    # 中文括号（）
    elif i.find('（')>=0 and i.find('）')>=0:
        print(i,i.find('（'),i.find('）'))
        drug_list.append(i)
```

降糖药的本名和别名分开存入2个列表

```python
drug_list1 = []
drug_list2 = []

for i in drug_list:
    if(i.find('(')>=0):
        if(len(i.split('('))==3):
            j, _, k = i.split('(')
        else:
            j, k = i.split('(')
        print(i.split('('))
    else:
        print(i.split('（'))
        if(len(i.split('（'))==3):
            j, _, k = i.split('（')
        else:
            j, k = i.split('（')
    drug_list1.append(j)
    drug_list2.append(k[:len(k)-1])
print(drug_list1)
print(drug_list2)
```

检测各药名包含的其他药名子串

```python
#   标注含可能误匹配子串drug_list1中药品
overlap_list1 = []
#   标注含可能误匹配子串drug_list2中药品
overlap_list2 = []
for i in range(len(drug_list)):
    # 列表存子串drug_list1中药品下标
    overlap1 = []
    # 列表存子串drug_list2中药品下标
    overlap2 = []
    for j in range(len(drug_list)):
        if i!=j:
            if drug_list1[i].find(drug_list1[j])>=0:
                overlap1.append(j)
            if drug_list2[i].find(drug_list2[j])>=0:
                overlap2.append(j)
    overlap_list1.append(overlap1)
    overlap_list2.append(overlap2)

print(overlap_list1)
print(overlap_list2)
                
```

三、从各病人信息中抽取各降糖药带药情况

```python
med_list = []
overlap_record = np.zeros(len(med_data))
num=0
for i in med_data:
    #   一维数组存25个药品有无的布尔值
    checklist = np.zeros(len(drug_list))

    #   find()字符串匹配查询drug_list1中药品名称
    for n in range(len(drug_list)):
        if i.find(drug_list1[n])>=0:
            checklist[n]=1
        
    #   排除含重合字符串的误匹配的drug_list1中药品
    for j in range(len(overlap_list1)):
        # 找出含可能误匹配子串drug_list1中药品下标
        if len(overlap_list1[j])!=0:
            # 如果查询到该药品
            if checklist[j]==1:
                # 删除该字符串，重新查询可能误匹配子串
                trunc = i.replace(drug_list1[j],'')
                for k in range(len(overlap_list1[j])):
                    # 若查询不到，则该药品置零
                    if(trunc.find(drug_list1[k])<0):
                        checklist[k]=0
                        overlap_record[num]=1
    
    #   find()字符串匹配查询drug_list1中药品名称
    for n in range(len(drug_list)):
        if i.find(drug_list2[n])>=0:
            checklist[n]=1
            
        #   排除含重合字符串的误匹配的drug_list2中药品
    for j in range(len(overlap_list2)):
        # 找出含可能误匹配子串drug_list2中药品下标
        if len(overlap_list2[j])!=0:
            # 如果查询到该药品
            if checklist[j]==1:
                # 删除该字符串，重新查询可能误匹配子串
                trunc = i.replace(drug_list2[j],'')
                for k in range(len(overlap_list2[j])):
                    # 若查询不到，则该药品置零
                    if(trunc.find(drug_list2[k])<0):
                        checklist[k]=0
                        overlap_record[num]=1
    num+=1
    med_list.append(checklist)
print(med_list)
print(overlap_record)
```

以表示有无带药的布尔值元素组成的数组记录每个病人的带药情况

```python
med_array = np.array(med_list).astype(int)
print(med_array.shape)
med_array = med_array.reshape([len(drug_list),len(med_data)])
print(med_array.shape)
df = pd.DataFrame({'Med0':med_array[0],'Med1':med_array[1],'Med2':med_array[2],'Med3':med_array[3],'Med4':med_array[4],
                   'Med5':med_array[5],'Med6':med_array[6],'Med7':med_array[7],'Med8':med_array[8],'Med9':med_array[9],
                   'Med10':med_array[10],'Med11':med_array[11],'Med12':med_array[12],'Med13':med_array[13],'Med14':med_array[14],
                   'Med15':med_array[15],'Med16':med_array[16],'Med17':med_array[17],'Med18':med_array[18],'Med19':med_array[19],
                   'Med20':med_array[20],'Med21':med_array[21],'Med22':med_array[22],'Med23':med_array[23],'Med24':med_array[24],
                   'Med25':med_array[25],'Med26':med_array[26],'Med27':med_array[27],
                   'Overlap':overlap_record.astype(int)})
print(df)
```
